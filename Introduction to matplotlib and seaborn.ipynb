# Barplot

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head(3)
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
s=df.species.value_counts()
s
```

```python
plt.bar(x=s.index, height=s.values)
```

```python
plt.bar(x=s.index, height=s.values)
plt.xlabel("Species")
plt.ylabel("Frequency/Count")
plt.title("Growth by month")
plt.show()
```

```python
plt.bar(x=s.index, height=s.values, color=['blue', 'green', 'maroon'])
plt.xlabel("Species")
plt.ylabel("Frequency/Count")
plt.title("Growth by month")
plt.show()
```

```python
sns.countplot(x='species', data=df) # barplot of a column
```

# Pie chart


```python
s=df.species.value_counts()

plt.pie(labels=s.index, x=s.values)
plt.show()
```

```python
s=df.species.value_counts()

plt.pie(labels=s.index, x=s.values, autopct="%.2f%%",colors=['maroon','green','grey'])
plt.show()
```
